```python
import matplotlib.pyplot as plt
import numpy as np

import pygmm

%matplotlib inline
```

```python
scenario = pygmm.Scenario(mag=6, dist_rup=20, site_cond="soil")
```

```python
as96 = pygmm.AbrahamsonSilva1996(scenario)
as96.interp(0.75)
```

```python
nai = np.linspace(0, 1)
stds = [-2, 0, 2]
```

```python
durs = as96.interp(nai, stds)
```

```python
fig, ax = plt.subplots()

for std, _durs in zip(stds, durs):
    ax.plot(_durs, nai, label=f"{std:.1f}")

ax.legend()
ax.set(xlabel="Duration (sec)", ylabel="Normalized $I_a$")
fig.tight_layout()
fig;
```
