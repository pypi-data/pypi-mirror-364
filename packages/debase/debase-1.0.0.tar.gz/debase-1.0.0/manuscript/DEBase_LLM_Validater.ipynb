# Test all the downloaded papers against the LLM extractor

Command: 
```
debase --manuscript "downloaded_papers/Stereospecific Enzymatic Conversion of Boronic Acids to Amines.pdf" --si "downloaded_papers/si-Stereospecific Enzymatic Conversion of Boronic Acids to Amines.pdf" --output "LLM/Stereospecific Enzymatic Conversion of Boronic Acids to Amines.csv" --keep-intermediates
```

````python
import time
import os 

data_dir = 'downloaded_papers/'
files = [f for f in os.listdir(data_dir) if f[:3] != 'si-']

with open('time_log.txt', 'w+') as fout:
    for i, f in enumerate(files):
        start = time.time()   
        output = f.replace('.pdf', '.csv')
        os.system(f'mkdir LLM/manuscript_{i}/')
        cmd = f'debase --manuscript "downloaded_papers/{f}" --si "downloaded_papers/si-{f}" --output "LLM/manuscript_{i}/{output}" --keep-intermediates'
        os.system(cmd)
        end = time.time()
        
        print('TIME:', end - start)
        fout.write(f'{f}\tmanuscript_{i}\t{end-start}\t{cmd}\n')
````
