- **‡πÑ‡∏ó‡∏¢** (`language='th'`) - ‡∏ä‡∏∑‡πà‡∏≠‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÄ‡∏õ‡πá‡∏ô‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢

````python
# ‡∏ô‡∏≥‡πÄ‡∏Ç‡πâ‡∏≤ ThaiFin Library
from thaifin import Stock, Stocks

# ‡∏ô‡∏≥‡πÄ‡∏Ç‡πâ‡∏≤ Library ‡πÄ‡∏™‡∏£‡∏¥‡∏°
from IPython.display import display
import pandas as pd
import warnings

warnings.filterwarnings("ignore")

print("‚úÖ ‡∏ô‡∏≥‡πÄ‡∏Ç‡πâ‡∏≤ ThaiFin Library ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")
print("üìä ‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡πÅ‡∏•‡πâ‡∏ß")
````

## üìã ‡∏Ñ‡∏•‡∏≤‡∏™ `Stocks` - ‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Å‡∏±‡∏ö‡∏´‡∏∏‡πâ‡∏ô‡∏´‡∏•‡∏≤‡∏¢‡∏ï‡∏±‡∏ß

### 1. üîç ‡∏Å‡∏≤‡∏£‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏´‡∏∏‡πâ‡∏ô (Smart Search)
‡∏à‡∏∞‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏≤‡∏°‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡πà‡∏≠ (symbol ‡∏´‡∏∏‡πâ‡∏ô) ‡∏´‡∏£‡∏∑‡∏≠‡∏ä‡∏∑‡πà‡∏≠‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó

```python
from thaifin import Stocks
stocks = Stocks(language='th')  # ‡πÉ‡∏ä‡πâ‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢
results = stocks.search('PTT')  # ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏´‡∏∏‡πâ‡∏ô PTT
for stock in results:
    print(f"‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô: {stock.name}, ‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡πà‡∏≠: {stock.symbol}")

````python
print("=== üîç ‡∏Å‡∏≤‡∏£‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏´‡∏∏‡πâ‡∏ô ===")
print("")

# ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô (‡∏†‡∏≤‡∏©‡∏≤‡∏≠‡∏±‡∏á‡∏Å‡∏§‡∏©)
print("1. ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤ 'cp' (‡∏à‡∏∞‡∏û‡∏ö‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡∏µ‡πà‡∏¢‡∏ß‡∏Ç‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ö CP):")
cp_stocks: list[Stock] = Stocks.search("cp", limit=3)
for stock in cp_stocks:
    print(f"   üìà {stock.symbol}: {stock.company_name}")

print("\n" + "=" * 50)

# ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢ (‡∏£‡∏∞‡∏ö‡∏ö‡∏à‡∏∞‡∏ï‡∏£‡∏ß‡∏à‡∏à‡∏±‡∏ö‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥)
print("2. ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤ '‡∏ò‡∏ô‡∏≤‡∏Ñ‡∏≤‡∏£' (‡∏à‡∏∞‡∏û‡∏ö‡∏ò‡∏ô‡∏≤‡∏Ñ‡∏≤‡∏£‡∏ï‡πà‡∏≤‡∏á‡πÜ):")
bank_stocks: list[Stock] = Stocks.search("‡∏ò‡∏ô‡∏≤‡∏Ñ‡∏≤‡∏£", limit=5)
for stock in bank_stocks:
    print(f"   üè¶ {stock.symbol}: {stock.company_name}")

print("\n" + "=" * 50)

# ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡∏≥‡∏†‡∏≤‡∏©‡∏≤‡∏≠‡∏±‡∏á‡∏Å‡∏§‡∏©
print("3. ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤ 'energy':")
energy_stocks: list[Stock] = Stocks.search("energy", limit=3)
for stock in energy_stocks:
    print(f"   ‚ö° {stock.symbol}: {stock.company_name}")
````

### 2. üìÉ ‡∏Å‡∏≤‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î

````python
print("=== üìÉ ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ===")
print("")

# ‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡πà‡∏≠‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
all_symbols = Stocks.list()
print(f"üìä ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏´‡∏∏‡πâ‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {len(all_symbols)} ‡∏ï‡∏±‡∏ß")
print(f"üî§ ‡∏´‡∏∏‡πâ‡∏ô 10 ‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å: {all_symbols[:10]}")

print("\n" + "=" * 50)

# ‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î (‡∏†‡∏≤‡∏©‡∏≤‡∏≠‡∏±‡∏á‡∏Å‡∏§‡∏©)
print("üìã ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏´‡∏∏‡πâ‡∏ô 5 ‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î (‡∏†‡∏≤‡∏©‡∏≤‡∏≠‡∏±‡∏á‡∏Å‡∏§‡∏©):")
stocks_df_en: pd.DataFrame = Stocks.list_with_names(language="en")
display(stocks_df_en.head())

print("\n" + "=" * 50)

# ‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î (‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢)
print("üìã ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏´‡∏∏‡πâ‡∏ô 5 ‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏£‡∏≤‡∏¢‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î (‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢):")
stocks_df_th: pd.DataFrame = Stocks.list_with_names(language="th")
display(stocks_df_th.head())
````

### 3. üè≠ ‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏≤‡∏°‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏≠‡∏∏‡∏ï‡∏™‡∏≤‡∏´‡∏Å‡∏£‡∏£‡∏°

````python
print("=== üè≠ ‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏≤‡∏°‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏≠‡∏∏‡∏ï‡∏™‡∏≤‡∏´‡∏Å‡∏£‡∏£‡∏° ===")
print("")

# ‡∏î‡∏π‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏≠‡∏∏‡∏ï‡∏™‡∏≤‡∏´‡∏Å‡∏£‡∏£‡∏°‡∏ó‡∏µ‡πà‡∏°‡∏µ
df_all: pd.DataFrame = Stocks.list_with_names(language="th")
sectors: pd.Series = df_all["sector"].value_counts()
print("üè≠ ‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏≠‡∏∏‡∏ï‡∏™‡∏≤‡∏´‡∏Å‡∏£‡∏£‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î:")
for sector, count in sectors.head(10).items():
    print(f"   ‚Ä¢ {sector}: {count} ‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó")

print("\n" + "=" * 50)

# ‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏ò‡∏ô‡∏≤‡∏Ñ‡∏≤‡∏£
banking_stocks: list[str] = Stocks.filter_by_sector("Banking")
print(f"üè¶ ‡∏´‡∏∏‡πâ‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏ò‡∏ô‡∏≤‡∏Ñ‡∏≤‡∏£ ({len(banking_stocks)} ‡∏ï‡∏±‡∏ß):")
print(f"   {banking_stocks}")

print("\n" + "=" * 30)

# ‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏û‡∏•‡∏±‡∏á‡∏á‡∏≤‡∏ô
energy_stocks: list[str] = Stocks.filter_by_sector("Energy")
print(f"‚ö° ‡∏´‡∏∏‡πâ‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏û‡∏•‡∏±‡∏á‡∏á‡∏≤‡∏ô ({len(energy_stocks)} ‡∏ï‡∏±‡∏ß):")
print(f"   {energy_stocks[:10]} ... (‡πÅ‡∏•‡∏∞‡∏≠‡∏∑‡πà‡∏ô‡πÜ)")
````

### 4. üèõÔ∏è ‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏≤‡∏°‡∏ï‡∏•‡∏≤‡∏î

````python
print("=== üèõÔ∏è ‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏≤‡∏°‡∏ï‡∏•‡∏≤‡∏î ===")
print("")

# ‡∏î‡∏π‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏´‡∏∏‡πâ‡∏ô‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏•‡∏≤‡∏î
df_all: pd.DataFrame = Stocks.list_with_names()
markets: pd.Series = df_all["market"].value_counts()
print("üèõÔ∏è ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏´‡∏∏‡πâ‡∏ô‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏•‡∏≤‡∏î:")
for market, count in markets.items():
    print(f"   ‚Ä¢ {market}: {count} ‡∏ï‡∏±‡∏ß")

print("\n" + "=" * 50)

# ‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô SET
set_stocks: list[str] = Stocks.filter_by_market("SET")
print(f"üìà ‡∏´‡∏∏‡πâ‡∏ô SET ({len(set_stocks)} ‡∏ï‡∏±‡∏ß):")
print(f"   ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: {set_stocks[:10]} ... (‡πÅ‡∏•‡∏∞‡∏≠‡∏∑‡πà‡∏ô‡πÜ)")

print("\n" + "=" * 30)

# ‡∏Å‡∏£‡∏≠‡∏á‡∏´‡∏∏‡πâ‡∏ô mai
mai_stocks: list[str] = Stocks.filter_by_market("mai")
print(f"üå± ‡∏´‡∏∏‡πâ‡∏ô mai ({len(mai_stocks)} ‡∏ï‡∏±‡∏ß):")
print(f"   ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: {mai_stocks[:10]} ... (‡πÅ‡∏•‡∏∞‡∏≠‡∏∑‡πà‡∏ô‡πÜ)")
````

## üè¢ ‡∏Ñ‡∏•‡∏≤‡∏™ `Stock` - ‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Å‡∏±‡∏ö‡∏´‡∏∏‡πâ‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏î‡∏µ‡∏¢‡∏ß

### 1. üìä ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó

````python
print("=== üè¢ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó ===")
print("")

# ‡∏™‡∏£‡πâ‡∏≤‡∏á Stock object (‡∏†‡∏≤‡∏©‡∏≤‡∏≠‡∏±‡∏á‡∏Å‡∏§‡∏©)
stock_en: Stock = Stock("PTT", language="en")
print("üá∫üá∏ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó PTT (‡∏†‡∏≤‡∏©‡∏≤‡∏≠‡∏±‡∏á‡∏Å‡∏§‡∏©):")
print(f"   üìù ‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡πà‡∏≠: {stock_en.symbol}")
print(f"   üè¢ ‡∏ä‡∏∑‡πà‡∏≠‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó: {stock_en.company_name}")
print(f"   üè≠ ‡∏≠‡∏∏‡∏ï‡∏™‡∏≤‡∏´‡∏Å‡∏£‡∏£‡∏°: {stock_en.industry}")
print(f"   üìä ‡∏Å‡∏•‡∏∏‡πà‡∏°: {stock_en.sector}")
print(f"   üèõÔ∏è ‡∏ï‡∏•‡∏≤‡∏î: {stock_en.market}")
print(f"   üåê ‡πÄ‡∏ß‡πá‡∏ö‡πÑ‡∏ã‡∏ï‡πå: {stock_en.website}")

print("\n" + "=" * 50)

# ‡∏™‡∏£‡πâ‡∏≤‡∏á Stock object (‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢)
stock_th: Stock = Stock("PTT", language="th")
print("üáπüá≠ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó PTT (‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢):")
print(f"   üìù ‡∏ä‡∏∑‡πà‡∏≠‡∏¢‡πà‡∏≠: {stock_th.symbol}")
print(f"   üè¢ ‡∏ä‡∏∑‡πà‡∏≠‡∏ö‡∏£‡∏¥‡∏©‡∏±‡∏ó: {stock_th.company_name}")
print(f"   üè≠ ‡∏≠‡∏∏‡∏ï‡∏™‡∏≤‡∏´‡∏Å‡∏£‡∏£‡∏°: {stock_th.industry}")
print(f"   üìä ‡∏Å‡∏•‡∏∏‡πà‡∏°: {stock_th.sector}")
print(f"   üèõÔ∏è ‡∏ï‡∏•‡∏≤‡∏î: {stock_th.market}")
print(f"   üåê ‡πÄ‡∏ß‡πá‡∏ö‡πÑ‡∏ã‡∏ï‡πå: {stock_th.website}")
````

### ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á üìà ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏ó‡∏£‡∏ô‡∏î‡πå‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï

````python
print("=== üìà ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏ó‡∏£‡∏ô‡∏î‡πå‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï ===")
print("")

# ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï‡∏Ç‡∏≠‡∏á PTT
ptt_stock = Stock("PTT", language="th")
print(f"üè¢ ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå: {ptt_stock.company_name}")

try:
    # ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏õ‡∏µ
    yearly_data: pd.DataFrame = ptt_stock.yearly_dataframe

    if not yearly_data.empty:
        # ‡πÉ‡∏ä‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 5 ‡∏õ‡∏µ‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô (‡∏´‡∏£‡∏∑‡∏≠‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏ñ‡πâ‡∏≤‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 5 ‡∏õ‡∏µ)
        recent_data: pd.DataFrame = yearly_data.tail(5)
        print(f"üìÖ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå: {len(recent_data)} ‡∏õ‡∏µ")
        print(f"üìä ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤: {recent_data.index.min()} ‡∏ñ‡∏∂‡∏á {recent_data.index.max()}")

        # Helper function to safely convert to number
        def safe_convert_to_number(value):
            if pd.isna(value) or value == "" or value is None:
                return None
            try:
                return float(str(value).replace(",", ""))
            except (ValueError, TypeError):
                return None

        # ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ
        if "‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏£‡∏ß‡∏°" in recent_data.columns:
            revenue_series: pd.Series = recent_data["‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏£‡∏ß‡∏°"]
            revenue_data: dict = {}
            for year, value in revenue_series.items():
                num_value: float | None = safe_convert_to_number(value)
                if num_value is not None:
                    revenue_data[year] = num_value

            if revenue_data:
                print("\nüí∞ ‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏£‡∏ß‡∏° (‡∏õ‡∏µ‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î):")
                for year, revenue in revenue_data.items():
                    print(f"   {year}: {revenue:,.0f} ‡∏•‡πâ‡∏≤‡∏ô‡∏ö‡∏≤‡∏ó")

                # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 2 ‡∏õ‡∏µ)
                if len(revenue_data) >= 2:
                    revenue_values: list[float] = list(revenue_data.values())
                    first_year_revenue: float = revenue_values[0]
                    last_year_revenue: float = revenue_values[-1]
                    years_diff: int = len(revenue_values) - 1

                    if first_year_revenue > 0:  # ‡∏´‡∏•‡∏µ‡∏Å‡πÄ‡∏•‡∏µ‡πà‡∏¢‡∏á‡∏Å‡∏≤‡∏£‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢‡∏®‡∏π‡∏ô‡∏¢‡πå
                        growth_rate: float = (
                            (last_year_revenue / first_year_revenue) ** (1 / years_diff)
                            - 1
                        ) * 100
                        print(f"   üìä ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {growth_rate:.2f}% ‡∏ï‡πà‡∏≠‡∏õ‡∏µ")

                        # ‡πÅ‡∏™‡∏î‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÅ‡∏õ‡∏•‡∏á‡∏£‡∏ß‡∏°
                        total_change: float = (
                            (last_year_revenue / first_year_revenue) - 1
                        ) * 100
                        print(
                            f"   üìà ‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÅ‡∏õ‡∏•‡∏á‡∏£‡∏ß‡∏°: {total_change:.2f}% ‡πÉ‡∏ô {years_diff} ‡∏õ‡∏µ"
                        )

        # ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï‡∏Ç‡∏≠‡∏á‡∏Å‡∏≥‡πÑ‡∏£
        if "‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥" in recent_data.columns:
            profit_series: pd.Series = recent_data["‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥"]
            profit_data: dict = {}
            for year, value in profit_series.items():
                num_value: float | None = safe_convert_to_number(value)
                if num_value is not None:
                    profit_data[year] = num_value

            if profit_data:
                print("\nüíµ ‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥ (‡∏õ‡∏µ‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î):")
                for year, profit in profit_data.items():
                    print(f"   {year}: {profit:,.0f} ‡∏•‡πâ‡∏≤‡∏ô‡∏ö‡∏≤‡∏ó")

                # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏¥‡∏ö‡πÇ‡∏ï‡∏Ç‡∏≠‡∏á‡∏Å‡∏≥‡πÑ‡∏£
                if len(profit_data) >= 2:
                    profit_values: list[float] = list(profit_data.values())
                    first_profit: float = profit_values[0]
                    last_profit: float = profit_values[-1]
                    if first_profit != 0:  # ‡∏´‡∏•‡∏µ‡∏Å‡πÄ‡∏•‡∏µ‡πà‡∏¢‡∏á‡∏Å‡∏≤‡∏£‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢‡∏®‡∏π‡∏ô‡∏¢‡πå
                        profit_change: float = ((last_profit / first_profit) - 1) * 100
                        print(f"   üíπ ‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÅ‡∏õ‡∏•‡∏á‡∏Å‡∏≥‡πÑ‡∏£: {profit_change:.2f}%")

        # ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå ROE
        if "ROE (%)" in recent_data.columns:
            roe_series: pd.Series = recent_data["ROE (%)"]
            roe_data: dict = {}
            for year, value in roe_series.items():
                num_value: float | None = safe_convert_to_number(value)
                if num_value is not None:
                    roe_data[year] = num_value

            if roe_data:
                print("\nüìä ROE (‡∏õ‡∏µ‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î):")
                for year, roe in roe_data.items():
                    print(f"   {year}: {roe:.2f}%")

                if len(roe_data) >= 2:
                    roe_values: list[float] = list(roe_data.values())
                    avg_roe: float = sum(roe_values) / len(roe_values)
                    max_roe: float = max(roe_values)
                    min_roe: float = min(roe_values)
                    print(f"   üìà ROE ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {avg_roe:.2f}%")
                    print(f"   üî∫ ROE ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î: {max_roe:.2f}%")
                    print(f"   üîª ROE ‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î: {min_roe:.2f}%")

        # ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏°: ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥
        if "‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥ (%)" in recent_data.columns:
            npm_series: pd.Series = recent_data["‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥ (%)"]
            npm_data: dict = {}
            for year, value in npm_series.items():
                num_value: float | None = safe_convert_to_number(value)
                if num_value is not None:
                    npm_data[year] = num_value

            if npm_data:
                npm_values: list[float] = list(npm_data.values())
                avg_npm: float = sum(npm_values) / len(npm_values)
                print(f"\nüìä ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≥‡πÑ‡∏£‡∏™‡∏∏‡∏ó‡∏ò‡∏¥‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {avg_npm:.2f}%")

    else:
        print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏≤‡∏¢‡∏õ‡∏µ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏ó‡∏£‡∏ô‡∏î‡πå")

except Exception as e:
    print(f"‚ùå ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î: {str(e)}")
````
